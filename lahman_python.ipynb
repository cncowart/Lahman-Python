### Q1 
a. Find all players in the database who played at Vanderbilt University.

b. Create a list showing each player’s first and last names as well as the total salary they earned in the major leagues.

c. Sort this list in descending order by the total salary earned.

d. Which Vanderbilt player earned the most money in the majors?

```python
people_college = pd.merge(people, collegeplaying, 
                               left_on = 'playerid', right_on = 'playerid', 
                               how = 'inner')
people_college.head(3)
```

```python
college_players  = pd.merge(people_college, schools, 
                            left_on = 'schoolid', right_on = 'schoolid', 
                            how = 'inner')
college_players
```

```python
college_salaries  = pd.merge(college_players, salaries, 
                            left_on = 'playerid', right_on = 'playerid', 
                            how = 'inner')
college_salaries
```

```python
vandy = college_salaries[college_salaries['schoolname'] == 'Vanderbilt University']
vandy
```

```python
vandy = vandy[['namefirst', 'namelast', 'schoolname', 'salary']]
vandy
```

```python
vandy['full_name'] = vandy['namefirst'] + ' ' + vandy['namelast']
vandy = vandy[['schoolname', 'salary', 'full_name']]
vandy
```

```python
vandy_salary = vandy.groupby('full_name', as_index=False)['salary'].sum()
vandy_salary
```

```python
vandy_salary = vandy_salary.sort_values(by=['salary'], ascending = False)
vandy_salary
```

# Q2
Using the fielding table, group players into three groups based on their position: label players with position OF as "Outfield", those with position "SS", "1B", "2B", and "3B" as "Infield", and those with position "P" or "C" as "Battery". Determine the number of putouts made by each of these three groups in 2016.

```python
fielding = pd.read_sql("SELECT * FROM fielding;", con=engine)
fielding
```

```python
fielding['pos_group'] = ''
fielding
```

```python
po_players = fielding[['playerid', 'pos', 'po', 'pos_group']]
po_players
```

```python
def pos_group(x):
    if x in ('SS','2B','3B', '1B'):
        return 'Infield'
    elif x == 'P':
        return 'Battery'
    elif x == 'C':
        return 'Battery'
    else: return 'Outfield'
```

```python
po_players['pos_group'] =  po_players['pos'].apply(pos_group)
po_players
```

```python
po_posi = po_players.groupby('pos_group', as_index=False)['po'].sum()
po_posi
```
